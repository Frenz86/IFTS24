#pulire il testo e rimuovere le parole 'gargamella', 'doggy', 'word3'

```python
import re

text1 = "Hello, gargamella doggy don't 123 @!£$%' \''s    \n hi"
text2 = "456 World!"

def clean_test(text):

    ####lower
    text_cleaned = text.lower()
    pattern1 = r"\d+"
    text_cleaned = re.sub(pattern1,'',text_cleaned)
    pattern2 = r"gargamella"
    text_cleaned = re.sub(pattern2,'',text_cleaned)
    ####
    return text_cleaned

clean_test(text1)
```

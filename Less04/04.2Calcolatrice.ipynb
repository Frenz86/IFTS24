```
%%writefile formule.py

class Calcolatrice:
    """
    questa classe ci permette dinamicamente di optare quale operazione vogliamo effettuare
    """

    def somma(self,*args):
        """
        il metodo somma, somma in maniera consequenziale i valori inseriti
        """
        s = 0
        for i in args:
            s += i
        return s

    def sottrazione(self,*args):
        """
        il metodo sottrazione, sottrae in maniera consequenziale i valori inseriti
        """
        sott = args[0]
        for i in args[1:]:
            sott -= i
        return sott

    def moltiplicazione(self,*args):
        """
        il metodo moltiplicazione, moltiplica in maniera consequenziale i valori inseriti
        """
        m = 1
        for i in args:
            m *= i
        return m

    def divisione(self,*args):
        """
        il metodo divisione, divide in maniera consequenziale i valori inseriti
        """
        if 0 in args:
            return 'uno degli input è zero'
        div = args[0]
        for i in args[1:]:
            div /= i
        return div
```
